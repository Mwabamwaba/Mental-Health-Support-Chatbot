```python
from transformers import pipeline
import gradio as gr
```

```python
emotion_classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=True
)
```

```python
with gr.Blocks() as interface:
    gr.Markdown("# 🧠 Mental Health Support Bot")
    gr.Markdown("This bot detects emotions and gives gentle, supportive replies.")

    chatbox = gr.Textbox(label="Conversation", lines=20, interactive=False)
    user_input = gr.Textbox(label="Your Message")
    state = gr.State([])  # Keeps history

    def respond(user_message, chat_history):
        return chatbot_response(user_message, chat_history)

    submit_btn = gr.Button("Send")
    submit_btn.click(respond, inputs=[user_input, state], outputs=[chatbox, state])

interface.launch()
```
